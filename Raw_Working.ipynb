```python
for i in range(1, 544):
    responce = requests.get(f"https://api.themoviedb.org/3/movie/top_rated?api_key=31e14bdb02f7ad309ec885dc97078c31&language=en-US&page={i}")
    temp_df = pd.DataFrame(responce.json()['results'])[['id', 'title', 'release_date', 'overview', 'popularity', 'vote_average', 'vote_count']]
    df = df.append(temp_df, ignore_index=True)
    
```
